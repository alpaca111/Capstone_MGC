### Eliminating certain 'genres'

Some categories are not representative of a genre and others have minimal amount of data collected in proportion to the entire dataset. Delete the following entries :
- 'International'
- 'Old-Time / Historic'
- 'Spoken'
- 'Easy listening'

```python
df_classifier1_0_100 = df_classifier1_0_100[~df_classifier1_0_100.genres_top.isin(['International','Old-Time / Historic',
                                                                                   'Spoken','Easy Listening'])]
df_classifier1_0_100.reset_index(inplace=True)
df_classifier1_0_100.to_csv('/Volumes/Extreme SSD/CAPSTONE_DATA/saved/final_dfs/df_classifier1_0_100_trim.csv')
```

```python
fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(14,6))
x = df_classifier1_0_100.genres_top.value_counts().index
y = df_classifier1_0_100.genres_top.value_counts().values
sns.barplot(x = x, y = y, ax=ax)
ax.set_xticklabels(labels = x, rotation = 30)  
ax.set_title("Count for each genre")
fig.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=1,figsize=(15,8))
sns.scatterplot(x = df_classifier1_0_100.H_std_10, y= df_classifier1_0_100.P_std_10,
                hue = df_classifier1_0_100.genres_top,ax=ax, s=30, palette = 'rainbow')
#ax.set_title("Count for each genre")
ax.set_title("Percussive STD vs Harmonic STD on 10th element")
ax.set(ylim=(0, 25))
ax.set(xlim=(0, 60))
fig.tight_layout()
plt.show()
```
